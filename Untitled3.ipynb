```python
def score_transaction(tx):
    score = 0
    rules = []

    # R1: amount > 3000 → +3 pkt
    if tx.get('amount', 0) > 3000:
        score += 3
        rules.append('R1')

    # R2: kategoria elektrionika i amount > 1500 → +2 pkt
    if tx.get('category') == 'elektronika' and tx.get('amount', 0) > 1500:
        score += 2
        rules.append('R2')

    # R3: godzina < 6 → +2 pkt
    hour = tx.get('hour')
    if hour is not None and hour < 6:
        score += 2
        rules.append('R3')

    return score, rules

# Test
test_tx = {'tx_id': 'TX999', 'amount': 4500.0, 'category': 'elektronika',
           'timestamp': '2026-04-01T03:15:00', 'hour': 3}
score, rules = score_transaction(test_tx)
print(score, rules)
```
